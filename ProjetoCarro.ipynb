```python
tipo_carro = int(input("Indique o tipo do carro \n(Econômico = 1 \nSedan = 2 \nSUV = 3)"))
valor_dia = 0
valor_seg = 40
taxa = 75
gps = 12
bebe = 15
cadeira = 15
assento = 15
opc_total = 0 

opc = input("Deseja algum opcional no veículo?")
while opc == "Sim":
    opc1 = input("Selecione o opcional: \nGPS = 1 \nBebê conforto = 2 \nCadeira de Bebê = 3 \nAssento elevado = 4 \nNão = 5 \n")
        
    if opc1 == "1":
            opc_total = opc_total + gps
            print("\n",opc_total)
            
    elif opc1 == "2":
            opc_total = opc_total + bebe
            print(opc_total)
            
    elif opc1 == "3":
            opc_total = opc_total + cadeira
            print(opc_total)
            
    elif opc1 == "4":
            opc_total = opc_total + assento
            print(opc_total)
        
    else:
            opc = "Não"

tempo = int(input("Quantos dias de aluguel? "))
        
def calculo(tipo_carro, valor_dia, valor_seg, taxa, opc_total, tempo):
    
    if tipo_carro == 1:
        valor_dia = 30
       
    elif tipo_carro == 2:
        valor_dia = 45
        
    elif tipo_carro == 3:
         valor_dia = 60
    
    else:
        return False
    
    return (valor_dia + valor_seg) * tempo + taxa + opc_total
  
textao = "Tipo do carro: "+str(tipo_carro)+"\nValor total " + str(calculo(tipo_carro, valor_dia, valor_seg, taxa, opc_total, tempo)) 

def salvar():
    file = open("projeto.txt", "w") 
    file.write(textao)

         
    
salvar()
```
